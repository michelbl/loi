```python
sac_par_mot = {m:Counter() for m in mots_importants}
nb_par_mot = {m:0 for m in mots_importants}
```

```python
for mots in mots_pp:
    for i, mot in enumerate(mots):
        if mot in mots_importants:
            nb_par_mot[mot] += 1
            
            if i-2 >= 0:
                sac_par_mot[mot][mots[i-2]] += 1
            if i-1 >= 0:
                sac_par_mot[mot][mots[i-1]] += 1
            if i+1 < len(mots):
                sac_par_mot[mot][mots[i+1]] += 1
            if i+2 < len(mots):
                sac_par_mot[mot][mots[i+2]] += 1
```
