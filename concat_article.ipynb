```python
import psycopg2
import xml.etree.ElementTree as ElementTree
import sys
import re
```

```python
repertoire_racine = '/home/michel/jorf_plat/'
connection = psycopg2.connect(dbname='jorf', user='jorf', password='baba')
curseur = connection.cursor()
f = open(repertoire_racine + 'concat_article.txt', 'w')
```

```python
curseur.execute("select BLOC_TEXTUEL from article;")

i = 0
while True:
    resultat_requete = curseur.fetchone()
    if not resultat_requete:
        break
    contenu_avec_balises = resultat_requete[0]
    contenu_xml = ElementTree.fromstring(contenu_avec_balises)
    contenu_html = ElementTree.tostring(contenu_xml, encoding='unicode', method='html')
    contenu_texte = re.sub('<[^<]+?>', '\n', contenu_html) + '\n'
    f.write(contenu_texte)
    if i%10000 == 0:
        print(i)
    i += 1
    
print('done : ' + str(i))
```

```python
f.close()
curseur.close()
connection.close()
```
