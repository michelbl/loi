```python
import pickle
from collections import Counter
from matplotlib import pyplot as plt
import numpy as np
%matplotlib inline
```

```python
repertoire_racine = '/home/michel/jorf_plat/'
```

## Coupe en 2 le document original

```python
with open(repertoire_racine + 'liste_mots_article.txt', 'rb') as f:
    document = pickle.load(f)
```

```python
document_permute = np.random.permutation(document)
```

```python
del document
```

```python
moitie = int(np.floor((len(document)+1) / 2))
```

```python
document_moitie_1 = list(document_permute[:moitie])
```

```python
with open(repertoire_racine + 'moitie_1.txt', 'wb') as f:
    pickle.dump(document_moitie_1, f)
```

```python
compteur_mot_1 = Counter(document_moitie_1)
```

```python
with open(repertoire_racine + 'compteur_mot_1.txt', 'wb') as f:
    pickle.dump(compteur_mot_1, f)
```

```python
del document_moitie_1
```

```python
document_moitie_2 = list(document_permute[moitie:])
```

```python
with open(repertoire_racine + 'moitie_2.txt', 'wb') as f:
    pickle.dump(document_moitie_2, f)
```

```python
compteur_mot_2 = Counter(document_moitie_2)
```

```python
with open(repertoire_racine + 'compteur_mot_2.txt', 'wb') as f:
    pickle.dump(compteur_mot_2, f)
```

```python
del document_moitie_2
```

```python
del document_permute
```

## Courbe log-log

```python
nb_mot = len(compteur_mot_1)
```

```python
liste_mot = list(compteur_mot_1.keys())
frequence_1 = [compteur_mot_1[mot] for mot in liste_mot]
order = np.argsort(frequence_1)
liste_mot = [liste_mot[index] for index in order][::-1]
frequence_2 = [compteur_mot_2[mot] for mot in liste_mot]
```

```python
list(zip(frequence_2, liste_mot))
```

```python
log_rang = np.log(np.arange(1, nb_mot+1))
log_frequence = np.log(np.array(frequence_2)/nb_mot)
```

```python
x = np.arange(14)
y1 = 3.3-0.95*x
y2 = 8.5-1.66*x
```

```python
plt.plot(log_rang, log_frequence)
```
